

Strategy 3:

○ Sort sellers in ascending order

○ Sort buyers in descending order

○ Allocate if required VMs are available

○ Find a break-even point: Break-even point is the largest index i for which buyer_bid[i]>seller_bid[i]

○ For example in given figure all Si are buyers and all ri are sellers. So here 3rd index is break-even point.

● Price to paid by all buyers= bid of the buyer at break-even point (i.e. 4)

● Price to be paid to all sellers=bid of the seller at break-even point (i.e. 8)

```python
sell_array = [['csp1', 20, 10], ['csp2', 30, 15]]
# buy_array = [['b1', 3, 20], ['b2', 4, 18], ['b3', 5, 16], ['b4', 6, 21], ['b5', 7, 24], ['b6', 8, 21]]
buy_array = [['b1', 2, 20], ['b2', 3, 12]]
# sorting the array of seler in asceding | of buyer in desceding
sel_array_sorted = sorted(sell_array,key = lambda x:x[2])
buy_array_sorted = sorted(buy_array,key = lambda x:x[2],reverse=True)



# allocation
allocation_record = []

for i in buy_array_sorted:
    for j in sel_array_sorted:
        if i[1] <= j[1]:
            allocation_record.append([i[0],j[0],i[2],j[2],i[1]])
            j[1] = j[1] - i[1]
            break
        else:
            pass

print('the allocation detail of the auction is') 
print('[buyer , seller , buyers offer, sellers offer, no. of vms]')
for i in allocation_record:
    print(i)


#break point
pnt = 0
for i in range(len(sel_array_sorted)):
  if sel_array_sorted[i][-1] > buy_array_sorted[i][-1]:
    break
  else:
    pnt = pnt + 1

print('the break even point is ',pnt - 1)

sel_value = sel_array_sorted[pnt-1][2]
buy_value = buy_array_sorted[pnt-1][2]



# finding the avrage profit of the seler
exe_seler = []
sel_final = []
for i in allocation_record:
    if i[1] not in exe_seler:
        exe_seler.append(i[1])
        sel_final.append(i)

a = []
for i in sel_final:
    a.append(sel_value - i[3])
    # a.append((sel_value-i[3])*i[4])

print('the Average profit of seller is : ', sum(a)/len(a))

# finding the avrage profit of the buyer
buy_final = []
exe_buer = []
for i in allocation_record:
    if  i[0] not in exe_buer:
        exe_buer.append(i[0])
        buy_final.append(i)

b= []
for i in buy_final:
    b.append(i[2]-buy_value)
    # b.append((i[2]-buy_value)*i[4])

print('The Average profit of buyer is : ', sum(b)/len(b))

# counting the profit loss of the broker 
br_profit = []

for i in allocation_record:
  br_profit.append((buy_value*i[4])-(sel_value*i[4]))

print(br_profit)
print('Profit: ',sum(br_profit))

```
